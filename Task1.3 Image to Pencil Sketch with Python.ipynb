## Image to Pencil Sketch with Python:

### import library

```python
import cv2
import matplotlib.pyplot as plt 
%matplotlib inline
```

### Read image

```python
image = cv2.imread(r"C:\Users\HP\Desktop\Internships\Letsgrow more\Tasks\Tasks1\Wallpapaer.jpg")
# image
```

### Read image in RGB format

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (8,8))
plt.imshow(img_rgb)
plt.title("Image")
```

### Converting origional image to grayscale

```python
GrayImage = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.figure(figsize = (8,8))
plt.imshow(GrayImage)
plt.title("Gray image")
# GrayImage
```

### Inverting image

```python
inverted_image = 255 - GrayImage
plt.figure(figsize = (8,8))
plt.imshow(inverted_image)
plt.title("Inverting image")
```

### Blur image by  the Gaussian function

```python
blur = cv2.GaussianBlur(inverted_image , (21,21) , 0)
plt.figure(figsize = (8,8))
plt.imshow(blur)
plt.title("Blur image")
```

```python
inverted_blur = 255 - blur
plt.figure(figsize = (8,8))
pencil_sketch = cv2.divide(GrayImage, inverted_blur , scale = 256.0)
plt.imshow(pencil_sketch)
```

### Final sktech

```python
plt.figure(figsize = (8,8))
final_image = cv2.cvtColor(pencil_sketch , cv2.COLOR_RGB2BGR)
plt.imshow(final_image)
plt.title("Final Image")
plt.axis(False)
plt.show()
```
