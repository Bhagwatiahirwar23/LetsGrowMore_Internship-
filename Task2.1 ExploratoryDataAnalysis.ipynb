#### Top 10 countries with highest number of atteckes

```python
people_died = terrorism[['country_txt' , 'nkill']].groupby('country_txt').sum()
people_died.head(10)
```

```python
people_died = terrorism[['iyear' , 'nkill']].groupby('iyear').sum()
people_died.head(10)
```

```python
people_died.plot(kind = 'bar' , color = 'b' , figsize = (18,6))
plt.xlabel('Years')
plt.ylabel('Number of people died ')
plt.title(' Number of people died with respect to years')
```

```python
region = terrorism[['region_txt' , 'nkill']].groupby('region_txt').sum()
region
```

```python
region.plot(kind = 'bar' , color = 'r' , figsize = (18,6))
plt.xlabel('region')
plt.ylabel('Number of people died ')
plt.title(' Number of people died with respect to region')
```

```python
attack_type = terrorism[['attacktype1_txt' , 'nkill']].groupby('attacktype1_txt').sum()
attack_type
```

```python
attack_type.plot(kind = 'bar' , color = 'm' , figsize = (18,6))
plt.xlabel('attack_type')
plt.ylabel('Number of people died ')
plt.title(' Number of people died with respect to attack_type')
```

```python
weapon= terrorism[['weaptype1_txt' , 'nkill']].groupby('weaptype1_txt').sum()
weapon
```

```python
weapon.plot(kind = 'bar' , color = 'g' , figsize = (18,6))
plt.xlabel('weapon used')
plt.ylabel('Number of people died ')
plt.title(' Number of people died with respect to weapon used')
```
